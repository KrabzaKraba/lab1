**Выбранный датасет**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import figure
%matplotlib inline

df = pd.read_csv("./Bitcoin Price (2014-2023)_exported.csv")
df.head()
```

**Нахождение среднее значение и СКО**

```python
def print_statistics(df, columns):
    for col in columns:
        mean_value = np.mean(df[col])
        std_value = np.std(df[col])
        print(f"\033[1m{col}\033[0m")
        print(f"СКО = {std_value}")
        print(f"Среднее = {mean_value}\n")
print_statistics(df, ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
```

**Построение гистограмм распределения значений и определение есть ли выбросы**

```python
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

def plot_histograms_and_detect_outliers(df, columns):
    for col in columns:
        # Построение гистограммы
        plt.figure(figsize=(10, 5))
        plt.hist(df[col], bins=30, alpha=0.7, color='blue', edgecolor='black')
        plt.title(f'Гистограмма распределения для {col}')
        plt.xlabel(col)
        plt.ylabel('Частота')
        plt.grid(axis='y', alpha=0.75)
        plt.show()

        # Определение выбросов
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

        print(f'Для {col}:')
        print(f'Нижняя граница выбросов: {lower_bound}')
        print(f'Верхняя граница выбросов: {upper_bound}')
        print(f'Количество выбросов: {outliers.shape[0]}')
        print(f'Выбросы:\n{outliers[col]}\n')

# Вызов функции
plot_histograms_and_detect_outliers(df, columns)
```

**Определение корреляции между параметрами**

```python
df = pd.read_csv('./Bitcoin Price (2014-2023)_exported.csv')
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
print("Матрица корреляции:")
print(correlation_matrix)
```

Open, High, Low, Close и Adj Close: Все эти параметры имеют очень высокую положительную корреляцию (близкую к 1). Это ожидаемо, поскольку они представляют собой разные аспекты цен одного и того же актива в течение одного и того же временного периода. Параметр Volume имеет более низкую корреляцию с ценами. Это говорит о том, что объем торгов не так сильно коррелирует с ценами, как другие параметры. Однако он все еще имеет положительную корреляцию, что может указывать на то, что при увеличении цен на биткойн объем торгов также может увеличиваться, но не так явно.
